```python
#y''+ y = sinx + cosx
# 1/a:
x=var('x')
y=function('y')(x)
eqd=diff(y, x, 2) + y == sin(x) + cos(x)
sol=desolve(eqd, y)
sol
```

```python
_K1, _K2=var('_K1', '_K2')
soll1 = sol.substitute(_K1==0, _K2==0)
soll2 = sol.substitute(_K1==0, _K2==1)
soll3 = sol.substitute(_K1==0, _K2==-1)

plot([soll1, soll2, soll3], x, -2, 2)
```

```python
reset()
```

```python
#1/b:
#y'' - y = e^2x
x=var('x')
y=function('y')(x)
eqd=diff(y, x, 2) - y == e ^ (2 * x)
sol=desolve(eqd, y)
sol
```

```python
_K1, _K2=var('_K1', '_K2')
soll = sol.substitute(_K1==1, _K2==4)
plot(soll, x, -8, 8)
```

```python
_K1, _K2=var('_K1', '_K2')
soll = sol.substitute(_K1==1, _K2==4)
plot(soll, x, -8, 8)
```
